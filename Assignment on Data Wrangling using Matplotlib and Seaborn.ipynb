```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"C:\Users\HP\Downloads\Customers.csv")
dataset.head(10)
```

```python
dataset.columns
```

```python
#To remove the columns we don't need
x = ['Prefix', 'FirstName', 'LastName','Unnamed: 5', 'Unnamed: 6','Unnamed: 5', 'Unnamed: 6','Unnamed: 8','Unnamed: 17',
       'Unnamed: 18','BirthDate.1']
```

```python
dataset.drop(x,axis=1, inplace=True)
```

```python
dataset
```

```python
#Renaming the Unnamed:4 column to full name. To rename columns we use .rename(columns='')
dataset.rename(columns={'Unnamed: 4':'FullName'},inplace=True)
dataset.head()
```

```python
#To rename the entries in thre MaritalStatus,Gender, and HomeOwner columns, we use .replace
dataset.MaritalStatus = dataset.MaritalStatus.replace('M','Married')
dataset.MaritalStatus = dataset.MaritalStatus.replace('S','Single')
dataset.Gender = dataset.Gender.replace('M','Male')
dataset.Gender = dataset.Gender.replace('F','Female')
dataset.HomeOwner = dataset.HomeOwner.replace('Y','Yes')
dataset.HomeOwner = dataset.HomeOwner.replace('N','No')
```

```python
dataset.head()
```

```python
#checking the datatypes
dataset.info()
```

```python
#To convert the Birthdate to datetime format
dataset.BirthDate=pd.to_datetime(dataset.BirthDate, format='%d/%m/%Y')
```

```python
dataset
```

```python
#To convert the Annual Income column to an integer datatype, remove the $ using str.strip and remove , using str.replace
dataset.AnnualIncome=dataset.AnnualIncome.str.lstrip('$')
dataset.AnnualIncome=dataset.AnnualIncome.str.replace(',','')
#After removing the unwanted characters, convert the column to integer using casting
dataset.AnnualIncome=dataset.AnnualIncome.astype(int)
```

```python
dataset.info()
```

```python
#To extract the birthmonth from the BirthDate column and save it as a column
dataset['BirthMonth']=dataset.BirthDate.dt.month
dataset.head(3)
```

```python
#To extract the Age
#First extract the year of birth
dataset.BirthDate.dt.year
```

```python
#declare a variable named 'current year'
current_year = 2024
```

```python
#Subtract the year of birth from the current year to get the age
current_year - (dataset.BirthDate.dt.year)
```

```python
#Add a new column for Age into the dataset
dataset['Age']=current_year - (dataset.BirthDate.dt.year)
dataset.head(4)
```

```python
#To rename the entries in the BirthMonth column ie 1-January, 2-February, 3-March, etc.
dataset.BirthMonth=dataset.BirthMonth.replace(1,'January')
dataset.BirthMonth=dataset.BirthMonth.replace(2,'February')
dataset.BirthMonth=dataset.BirthMonth.replace(3,'March')
dataset.BirthMonth=dataset.BirthMonth.replace(4,'April')
dataset.BirthMonth=dataset.BirthMonth.replace(5,'May')
dataset.BirthMonth=dataset.BirthMonth.replace(6,'June')
dataset.BirthMonth=dataset.BirthMonth.replace(7,'July')
dataset.BirthMonth=dataset.BirthMonth.replace(8,'August')
dataset.BirthMonth=dataset.BirthMonth.replace(9,'September')
dataset.BirthMonth=dataset.BirthMonth.replace(10,'October')
dataset.BirthMonth=dataset.BirthMonth.replace(11,'November')
dataset.BirthMonth=dataset.BirthMonth.replace(12,'December')
```
